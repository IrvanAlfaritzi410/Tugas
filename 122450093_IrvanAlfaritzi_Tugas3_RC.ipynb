Mengimpor Library yang diperlukan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset

```python
# Memuat dataset Titanic yang disediakan oleh Seaborn
titanic = sns.load_dataset('titanic')

# Menampilkan 5 baris pertama dari data untuk memastikan data berhasil dimuat
titanic.head()
```

1. Visualisasi Tingkat Keselamatan Berdasarkan Jenis Kelamin

```python
# Set style untuk visualisasi
sns.set(style="whitegrid")

# Membuat visualisasi tingkat keselamatan berdasarkan jenis kelamin
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='survived', data=titanic, palette='viridis')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

2. Visualisasi Tingkat Keselamatan Berdasarkan Kelas Penumpang

```python
# Visualisasi tingkat keselamatan berdasarkan kelas penumpang
plt.figure(figsize=(10, 6))
sns.countplot(x='class', hue='survived', data=titanic, palette='viridis')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

3. Visualisasi Distribusi Usia Penumpang yang Selamat dan Tidak Selamat

```python
# Visualisasi distribusi usia penumpang yang selamat dan tidak selamat
plt.figure(figsize=(12, 6))
sns.histplot(data=titanic, x='age', hue='survived', element='step', stat='density', common_norm=False, palette='viridis')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
